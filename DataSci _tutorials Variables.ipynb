What is a Variable?
• A variable in Python is like a container that holds data.
• You can store different types of values in a variable, such as text, numbers,
logical or even more complex data structures like lists or dictionaries.
• Variables make it easy to reference and manipulate data throughout your
program.
• Dynamic Typing: In Python, you don’t need to declare the data type of a variable.
The type is automatically assigned based on the value you store in it.
• Variable Assignment: Use the = operator to assign a value to a variable
• Re-assigning Variables: Variables can be reassigned to hold different values, even
values of different types.
Declare a Python Variable - Golden Rules
1. Use Meaningful Names
Example: price, total_sales, student_name are better than p, ts, or sname.
2. Use Snake Case for Variable Names
In Python, the convention is to use snake_case for variable names (all lowercase with
underscores between words or/and numbers).
Example: total_revenue, average_temperature,student1, student_1
3. Avoid Reserved Keywords
Do not use Python's reserved keywords (e.g., class, for, if, etc.) as variable names, as it can
lead to errors.
Example: class = 10 will raise a syntax error. Use something like class_name instead.
4. Stick to Lowercase Letters (for most variables)
For general variables, stick to lowercase names. Uppercase letters are typically reserved
for constants.
Example: discount_rate, user_id.
5. Avoid Single Character Names (Unless They are Obvious)
Using single character variable names like x or i is acceptable in small contexts, such as
loops or mathematical functions. However, avoid them for more meaningful variables.
Example: for i in range(10) is fine, but x = "John" is unclear.
6. Use Constants Appropriately
Constants are typically written in ALL_CAPS. This helps other developers recognize that the
value should not be modified.
Example: PI = 3.14159, MAX_RETRIES = 5.
7. Do Not Start Variable Names with Digits
Variable names cannot begin with a number, but numbers can be used after the first letter.
Example: 2cars = 10 is invalid, but cars2 = 10 is fine.
8. Use Clear and Consistent Naming for Booleans
Booleans should be named in a way that clearly implies a True/False value, often prefixed
with is, has, or should.
Example: is_active, has_permission.
9. Avoid Unnecessary Global Variables
Limit the use of global variables. Instead, encapsulate variables within functions or classes
where possible to avoid potential conflicts.
10.Python is case sensitive.
Therefore think wisely when defining variable names.
Example. my_var (Recommended) and mY_var are different variables.
11.Do Not Use Mutable Objects as Default Arguments
When using mutable objects like lists or dictionaries as default arguments in functions,
they can lead to unexpected behavior. Instead, use None as a default and handle it within
the function.
Example: a_list = none
Common Data Types:
• int: Integer numbers (e.g., 5, -2, 100)
• float: Floating-point numbers or decimals (e.g., 3.14, -0.5)
• str: Strings, or sequences of characters (e.g., "Hello, Python!")
• bool: Boolean values (True or False)

```python
# 1. Basic variable assignment using the '=' operator
student_name = "Anuroop"
print(student_name)
```

```python
# 2. Integer data type (int)  whole numbers
semester = 1
print(semester)
```

```python
# 3. Float data type  decimal numbers
grade_average = 8.75
print(grade_average)
```

```python
# 4. Boolean with descriptive naming (is_ prefix)
is_enrolled = True
print(is_enrolled)
```

```python
# 5. Constants use ALL_CAPS by convention
MAX_CREDITS = 30
print(MAX_CREDITS)
```

```python
# 6. Snake_case for multi-word variables
total_courses_completed = 5
print(total_courses_completed)
```

```python
# 7. Dynamic typing  reassigning to different type
status = 200   # initially an integer
print(status)  
status = "active"  # now a string
print(status)
```

```python
# 8. Checking variable types with type()
print(type(student_name), type(semester), type(grade_average), type(is_enrolled))
```

```python
# 9. Case sensitivity demo - these are different variables
course_id = "DS101"
Course_ID = "DS102"
print(course_id, Course_ID)   
```

```python
# 10. Variables in expressions
remaining_credits = MAX_CREDITS - (total_courses_completed * 5)
print(f"{student_name} has {remaining_credits} credits remaining in Semester {semester}")
```
