```python
#1. Understanding program flow
#Python executes code ine line at a time from top to bottom

print("Hello, This is Python")
print("Start")
print("This is line 1")
print("This is line 2")
print("End")
```

```python
# 2. Read input from a user and use a variable to store that user input
# the input() function pauses execution and awaits user input
print("Hello, This is Python")
print("---------------------")
n1 = input("What is your name?")#Stores user's input in variable n1
print("Hello, nice to meet you!",n1) #prints greeting with the name stored through the input() function
print("Emjoy Learning Python")
```

```python

# 3. Utillizing more variables to read various types of user data
# Each input() call stores different user information

name = input("What is your name ?")
print("Hello, nice to meet you !",name)

age = input("How old are you?")
print("Cool you are",age,"years old")

height = input("How tall are you in meters?")
print("Your height is ",height,"meters")

location = input("Where are you right now?")
print("You have are currently in ",location)
```

```python
# 4. Understanding the input() functions behaviour wiht differnt data types 
# the input() fn always returns a string reagardless of what the user inputs so type casting can be done afterwords

print("Variables through Input")
print("Name :", name, type(name))        # <class 'str'>
print("Age :", age, type(age))            # <class 'str'> - even numbers
print("height :", height, type(height))  # <class 'str'> - decimals are also strings
print("Location:", location, type(location))  # <class 'str'>
```

```python

#5. Conversion of variables to specific data types 
#Use int(), float(), bool() to do data type conversion

print("Converted Variables: ")
age = int(age)
print("Age: ", age, type(age))

height = float(height)
print("Height:", height, type(height))

location = bool(location)
print("Location is in Berlin?",location, type(location))
```

```python

#6. Inline conversion of variables 
# Convert input directly while reading to increase efficiency 

name = input(" What is your name?")                         #input is mostly string 
age = int(input("How old are you ?"))                       #Inline conversion to int
height = float(input("How tall are you in meters?"))        #Inline conversion to float 
location = bool(input("Are you currently in Berlin?"))      #Inline conversion to bool

print("Hey, Nice to meet you!",name)
print("You are:",age,"Years old")
print("You are",height,"meters tall")
print("You are currently in Berlin?:",location)
```

```python
# Check the variable types after inline conversion
# Now each variable has its correct data type

#checking the variable types to understand inline conversion working 
#now each variable has its correct data type

print("Variables in the INPUT after inline conversion above:")
print("name:",name, type(name))                #<class 'str'>
print("age:",age,type(age))          #<class 'int'> 
print("height:",height, type(height))                       #<class 'float'>      
print("Location truth:",location,type(location) )         #<class 'bool'>   
```
